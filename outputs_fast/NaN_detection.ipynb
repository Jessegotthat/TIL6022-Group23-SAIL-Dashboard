```python
import pandas as pd
from pathlib import Path

# ========= æ–‡ä»¶è·¯å¾„ =========
# è¾“å…¥æ–‡ä»¶ï¼šä½ ç°åœ¨çš„è®­ç»ƒé›†ï¼ˆå« timestampã€ä¼ æ„Ÿå™¨ã€å¤–ç”Ÿå˜é‡ï¼‰
in_path  = Path("C:/Users/elvinli/OneDrive/CodeProjects/TIL6022-Group23-SAIL-Dashboard/outputs_fast/training_dataset_tomtom_vessel.csv")
# è¾“å‡ºæ–‡ä»¶ï¼šæ·»åŠ  _isna åˆ—åçš„ç‰ˆæœ¬
out_path = Path("C:/Users/elvinli/OneDrive/CodeProjects/TIL6022-Group23-SAIL-Dashboard/outputs_fast/training_dataset_tomtom_vessel_ready.csv")

# ========= 1. è¯»å–æ–‡ä»¶ =========
df = pd.read_csv(in_path, parse_dates=["timestamp"], low_memory=False)
print(f"âœ… å·²è¯»å–æ–‡ä»¶ï¼š{in_path}")
print("æ•°æ®å½¢çŠ¶ï¼š", df.shape)

# ========= 2. è‡ªåŠ¨æ£€æµ‹å¤–ç”Ÿå˜é‡åˆ— =========
# è‡ªåŠ¨åŒ¹é…åˆ—åä¸­åŒ…å« "tomtom" æˆ– "vessel" çš„åˆ—ï¼ˆä¸åŒºåˆ†å¤§å°å†™ï¼‰
exog_cols = [c for c in df.columns if any(k in c.lower() for k in ["tomtom", "vessel"])]

if not exog_cols:
    raise ValueError("âŒ æœªæ£€æµ‹åˆ°å¤–ç”Ÿå˜é‡åˆ—ï¼Œè¯·æ£€æŸ¥åˆ—åæ˜¯å¦åŒ…å« 'tomtom' æˆ– 'vessel'ã€‚")
else:
    print("ğŸ” æ£€æµ‹åˆ°å¤–ç”Ÿå˜é‡åˆ—ï¼š", exog_cols)

# ========= 3. ä¸ºæ¯ä¸ªå¤–ç”Ÿå˜é‡åˆ›å»º _isna ç¼ºå¤±æŒ‡ç¤ºåˆ— =========
for col in exog_cols:
    new_col = f"{col}_isna"
    df[new_col] = df[col].isna().astype("int8")

print(f"âœ… å·²ä¸º {len(exog_cols)} ä¸ªå¤–ç”Ÿå˜é‡ç”Ÿæˆç¼ºå¤±æŒ‡ç¤ºåˆ—ã€‚")

# ========= 4. è¾“å‡ºæ–‡ä»¶ =========
df.to_csv(out_path, index=False, float_format="%.6f")
print(f"âœ… å·²ä¿å­˜æ–‡ä»¶ï¼š{out_path}")
print("æ–°æ•°æ®å½¢çŠ¶ï¼š", df.shape)

# ========= 5. å¯é€‰ï¼šç®€å•éªŒè¯ =========
print("\nğŸ“Š ç¼ºå¤±æ¯”ä¾‹æ£€æŸ¥ï¼š")
for col in exog_cols:
    miss_rate = df[col].isna().mean() * 100
    print(f"{col:<35} ç¼ºå¤±ç‡ = {miss_rate:5.2f}%")
```
